```r
text = read_xlsx("/content/Master's_Speeches.xlsx")
```

```r
text$texto = text$texto %>%
  str_to_lower() %>%
  str_replace_all(" *-+ *", "") %>%
  str_replace_all("[[:punct:]]", "") %>%
  str_replace_all("ye", "") %>%
  str_replace_all("thou", "") %>%
  str_replace_all("thee", "") %>%
  str_replace_all("thy", "") %>%
  str_replace_all("thine", "") %>%
  removeNumbers() %>%
  trimws()
```

```r
text_df = text %>%
  unnest_tokens(word, texto)
```

```r
text_df
```

```r
tidy_df = text_df %>%
anti_join(stop_words)
```

```r
tidy_df
```

```r
tidy_df %>%
  count(word, sort = TRUE)
```

```r
tidy_df %>%
  count(word) %>%
  with(wordcloud(word, n, max.words = 900))
```
